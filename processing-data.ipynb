# Handle Mistakes in Data Entries for some files, the netmanager column was swapped by the purchase area.

```python
def correct_swapped_values(row):
    try:
        # Trying to convert 'net_manager' to a float. If it succeeds, it's likely a number.
        float(row['net_manager'])
        # If successful, swap 'net_manager' and 'purchase_area'
        return row['purchase_area'], row['net_manager']
    except ValueError:
        # If it fails, 'net_manager' is not a number and is likely correct
        return row['net_manager'], row['purchase_area']

# Apply the correction function to each row
df_train['net_manager'], df_train['purchase_area'] = zip(*df_train.apply(correct_swapped_values, axis=1))
df_test['net_manager'], df_test['purchase_area'] = zip(*df_test.apply(correct_swapped_values, axis=1))
```
